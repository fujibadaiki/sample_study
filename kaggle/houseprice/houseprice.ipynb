```
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # visualization library
import matplotlib.pyplot as plt # visualization library
import plotly.plotly as py # visualization library
from plotly.offline import init_notebook_mode, iplot # plotly offline mode
init_notebook_mode(connected=True) 
import plotly.graph_objs as go # plotly graphical object
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
```

```
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```
train.info()
```

```
test.info()
```

```
train = train.drop(['Utilities','BsmtFinType1','BsmtFinType2','PoolQC','GarageYrBlt'], axis=1)
test =  test.drop(['Utilities','BsmtFinType1','BsmtFinType2','PoolQC','GarageYrBlt'], axis=1)
```

```
train = train.fillna({'LotFrontage':0,'MasVnrType':'None','MasVnrArea':0,
                    'BsmtQual':'TA','BsmtCond':'TA','BsmtExposure':'No'})

test = test.fillna({'MSZoning':'RL','LotFrontage':0,'Exterior1st':'VinylSd','Exterior2ed':'VinylSd','MasVnrType':'None',
                  'MasVnrArea':0,'BsmtQual':'TA','BsmtCond':'TA','BsmtExposure':'No',
                 'BsmtFinSF1':0,'BsmtFinSF2':0,'BsmtUnfSF':0,'TotalBsmtSF':0,'BsmtFullBath':0,
                 'BsmtHalfBath':0,'KintchenQual':'TA','Functional':'Typ','GarageCars':3,'GarageArea':test['GarageArea'].mean(),'GarageCars':500,
                 'SaleType':'WD'})
```

```
train.shape
```

```
train = train.astype({'MSSubClass':'category'})
train = pd.get_dummies(train,dummy_na=True)
```

```
test = test.astype({'MSSubClass':'category'})
test = pd.get_dummies(test,dummy_na=True)
```

```
train = train.drop(['Condition2_RRAe','Condition2_RRAn','Condition2_RRNn',
                    'HouseStyle_2.5Fin','RoofMatl_ClyTile','RoofMatl_Membran','RoofMatl_Metal',
                   'RoofMatl_Roll','Exterior1st_ImStucc','Exterior1st_Stone','Exterior2nd_Other',
                   'Heating_Floor','Heating_OthW','Electrical_Mix','GarageQual_Ex','MiscFeature_TenC'], axis=1)
test =  test.drop(['MSSubClass_150.0'], axis=1)

```

```
from sklearn.preprocessing import MinMaxScaler
```

```
from sklearn.neural_network import MLPRegressor
X,y = train.loc[:,train.columns != 'SalePrice'], train.loc[:,'SalePrice']
scaler = MinMaxScaler()
scaler.fit(X)
train = scaler.transform(X)
X_train, X_test,y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```
X_train.isnull().sum()
```

```
model = MLPRegressor(max_iter=1000, random_state=0, alpha=1)
#model = MLPRegressor(hidden_layer_sizes=(100,100,100,100,),random_state=42)
model.fit(X_train, y_train)
print(model.score(X_test,y_test))
```

```
test.isnull().sum()
```

```
result=model.predict(test)
print(result)
```
