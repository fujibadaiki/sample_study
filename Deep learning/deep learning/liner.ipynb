```
import numpy as np
import matplotlib.pyplot as plt
```

```
n_clos=2
n_rows=2
```

```
plt.figure(figsize=(20,8))
```

```
x = np.arange(-5, 5, 0.1)
```

```
y = x
plt.subplot(n_rows,n_clos,1)
plt.title('x')
plt.ylim(-10,10);plt.plot(x,y)
```

```
y = -2 * x-3 * x+x
plt.subplot(n_rows,n_clos,2)
plt.title('2x-3x+x')
plt.ylim(-10,10);plt.plot(x,y)
```

```
y = np.sin(x)
plt.subplot(n_rows,n_clos,3)
plt.title('sin(x)')
plt.ylim(-10,10);plt.plot(x,y)
```

```
y = np.exp(x)
plt.subplot(n_rows,n_clos,4)
plt.title('exp(x)')
plt.ylim(-10,10);plt.plot(x,y)
```
